### How is Volatility Calculated?

The volatility of the second 10-minutes of each (10 minute, 10 minute) pair is calculated as follows:

$$
\sigma = \sqrt{\sum_{t}r_{t}^2}
$$

and for variance
$$
\sigma^2 = \sum_{t}r_{t}^2
$$

Where $r$ is the _log_return_ between the previous data point and the current one. Note that the _log_return_ cannot be calculated for the first data point in each hour, in which case the _log_return_ will be the special value _NaN_ (not-a-number).

Another useful relationship if we want to use different time intervals is the following:
$$
\sigma_{1}^{2} = \sigma_{2}^{2} \cdot \frac{t_{1}}{t_{2}}
$$

You don't need to use this relationship, as all the `target_vars` are 10 minute variances and the orderbook data preceeding it is 10minutes in length. However it becomes important if you start working with returns that don't sum to 10 minutes in duration.


#### Built-in Model

#### Step-by-step: What the baseline model does

The baseline model computes volatility for each 10 minute bucket i.e., each `(stock_id, time_id)` as:

1) Take all rows within that 10 minutes for a given stock
2) Drop `NaN` values in `log_return` (the first row has no previous price)
3) Compute the variance

````python
import pandas as pd
from typing import List

# This function declaration makes clear the inputs and outputs, structured as: 
# def func_name(arg1: type(arg1)) -> type(return value)
def _var_unweighted(group: pd.DataFrame) -> float:
    """Unweighted variance: sum of squared log returns.

    Returns the unweighted variance for a single (stock_id, time_id)
    """
    log_returns = group["log_return"].dropna()
    return float((log_returns ** 2).sum()) if not log_returns.empty else 0.0
````

Naturally we want to use `_var_unweighted` to calculate a variance for each `(stock_id, time_id)`. For this we need new tools, `groupby` and `apply`.


``` python

```

This splits the dataframe into groups, one group for each `(stock_id, time_id)` pair. After this step, `gdf` isn’t a normal dataframe anymore, it’s a special object that represents a collection of sub-dataframes, each with rows from a single stock and time bucket. Somewhat similar to a dictionary.

``` python

```

Here, apply runs the function `_var_unweighted` separately on each group.

- `grp` is just a normal dataframe containing the rows for one particular `(stock_id, time_id)`.
- `_var_unweighted(grp)` computes the variance for that group.
- `apply` collects all these results and returns a Series, where the index is `(stock_id, time_id)` and the values are the variances.

````python
import numpy as np

def _var_weighted(
    group: pd.DataFrame,
    decay: float = 0.01,                   # per-second decay rate
    weight_col: str = "seconds_in_bucket"
) -> float:
    """
    Weighted realized variance using backward exponential decay.
    Later observations get higher weight, but the weights are normalized
    to sum(len(returns)) so the output remains a "sum" scale.
    """
    log_returns = group["log_return"].dropna()
    if log_returns.empty:
        return 0.0

    t = group.loc[log_returns.index, weight_col].values.astype(float)
    t_max = t.max()
    distances = t_max - t                     # time since observation (0 = latest)
    raw_weights = np.exp(-decay * distances)  # high for recent, low for older
    N = len(log_returns)

    # Normalize: make sum of weights = N  (preserves total variance magnitude)
    w = raw_weights / raw_weights.sum() * N

    return float((w * (log_returns.values ** 2)).sum())


def predict_baseline(df, prediction_index, decay=0.01):
    gdf = df.groupby(["stock_id", "time_id"], sort=True)
    all_vars = gdf.apply(lambda grp: _var_weighted(grp, decay=decay), include_groups=False)
    return all_vars.loc[prediction_index]


def cost(c):
    predicted_vars = predict_baseline(df, known_prediction_index, c)
    return score(predicted_vars, target_vars, plot=True, log_axes=True)


x = np.arange(0.005, 0.006, 0.0001)
y = [(c, cost(c)) for c in x]
y_plt = [c[1] for c in y]
plt.plot(x, y_plt)
best_c, best_cost = min(y, key=lambda x: x[1])
print(best_c, best_cost)
````

````python
import numpy as np
import pandas as pd
from arch import arch_model

def fit_garch_models(df: pd.DataFrame) -> tuple[dict, float]:
    """
    Fit a GARCH(1,1) on the time‑ordered realized variance of each stock.
    Scales the variance UP so that the median value across all stocks is ~1.0.
    Returns (models, scale_factor) where scale_factor was used to multiply the data.
    """
    # 1. Compute historical realized variances (unweighted sum of squared returns)
    def _var_unweighted(group):
        log_returns = group["log_return"].dropna()
        return float((log_returns ** 2).sum()) if not log_returns.empty else 0.0

    rv = df.groupby(["stock_id", "time_id"], sort=True).apply(
        _var_unweighted, include_groups=False
    )

    # 2. Determine a scaling factor: make the median variance ~ 1.0
    nonzero_rv = rv[rv > 1e-20]
    if nonzero_rv.empty:
        scale_factor = 1e11      # fallback if all values are zero
    else:
        median_var = nonzero_rv.median()
        # We want scaled_var = median_var * scale_factor ≈ 1.0
        scale_factor = 1.0 / median_var if median_var > 0 else 1e11

    # 3. Fit GARCH per stock
    models = {}
    for stock_id, series in rv.groupby(level="stock_id"):
        series_sorted = series.sort_index(level="time_id")
        # Remove zeros/near‑zeros
        series_clean = series_sorted[series_sorted > 1e-20]

        if len(series_clean) < 30:
            models[stock_id] = None
            continue

        # Multiply (NOT divide) to bring values into 1‑1000 range
        scaled_var = series_clean.values * scale_factor

        try:
            model = arch_model(scaled_var, mean="Zero", vol="GARCH", p=1, q=1, dist="normal")
            res = model.fit(disp="off", show_warning=False)
            models[stock_id] = res
        except Exception:
            models[stock_id] = None

    return models, scale_factor
````

````python
def predict_garch(
    models: dict,
    prediction_index: pd.MultiIndex,
    scale_factor: float,
    df: pd.DataFrame = None          # fallback (optional)
) -> pd.Series:
    """
    Generate forecasts for required (stock_id, time_id).
    The prediction is unscaled by dividing by the same scale_factor.
    """
    preds = []
    for stock_id, time_id in prediction_index:
        res = models.get(stock_id)
        if res is not None:
            forecast = res.forecast(horizon=1)
            scaled_forecast = forecast.variance.values[-1, 0]
            variance_forecast = scaled_forecast / scale_factor   # reverse multiplication
        else:
            # Optional fallback: use current bucket unweighted variance
            if df is not None:
                group = df[(df["stock_id"] == stock_id) & (df["time_id"] == time_id)]
                variance_forecast = _var_unweighted(group)
            else:
                variance_forecast = np.nan
        preds.append(variance_forecast)

    return pd.Series(preds, index=prediction_index)
````

````python
import numpy as np
import math

def _var_weighted(
    group: pd.DataFrame,
    decay: float = 0.01,               # per-second decay rate (higher = sharper recency)
    weight_col: str = "seconds_in_bucket"
) -> float:
    """
    Weighted realized variance (sum scale) with backward exponential weights.
    Weights: w_i = exp(-decay * (t_max - t_i)), then normalized so sum(w) = N.
    """
    log_returns = group["log_return"].dropna()
    if log_returns.empty:
        return 0.0

    t = group.loc[log_returns.index, weight_col].values.astype(float)
    t_max = t.max()
    distances = t_max - t                     # 0 for the very last observation
    raw_weights = np.exp(-decay * distances)  # exponential decay backward
    N = len(log_returns)

    # Normalize so that total weight = N (preserves total variance magnitude)
    w = raw_weights / raw_weights.sum() * N

    return float((w * (log_returns.values ** 2)).sum())

def predict_baseline_weighted(
    df: pd.DataFrame,
    prediction_index: pd.MultiIndex,
    decay: float = 0.01
) -> pd.Series:
    gdf = df.groupby(["stock_id", "time_id"], sort=True)
    all_vars = gdf.apply(
        lambda grp: _var_weighted(grp, decay=decay),
        include_groups=False
    )
    return all_vars.loc[prediction_index]


````

````python
import numpy as np
import pandas as pd

# ------------------------------------------------------------------
# 1. Helper – unweighted realized variance of a single bucket
# ------------------------------------------------------------------
def _var_unweighted(group: pd.DataFrame) -> float:
    log_returns = group["log_return"].dropna()
    return float((log_returns ** 2).sum()) if not log_returns.empty else 0.0

# ------------------------------------------------------------------
# 2. Compute realized variance for all (stock_id, time_id) we need
# ------------------------------------------------------------------
def compute_rv(df: pd.DataFrame, index: pd.MultiIndex) -> pd.Series:
    """Return realized variance for every (stock_id, time_id) in `index`."""
    # Filter df to only the requested index pairs
    df_filtered = df.set_index(["stock_id", "time_id"]).loc[index].reset_index()
    if df_filtered.empty:
        return pd.Series(dtype=float)
    rv = df_filtered.groupby(["stock_id", "time_id"], sort=True).apply(
        _var_unweighted, include_groups=False
    )
    return rv

# ------------------------------------------------------------------
# 3. Tune the EWMA smoothing parameter α on known targets
# ------------------------------------------------------------------
def tune_alpha(target_vars: pd.Series, rv_known: pd.Series) -> float:
    """
    Walk-forward EWMA one‑step‑ahead RMSE for a range of α.
    Returns the best α.
    """
    # rv_known is aligned with target_vars (same index)
    alphas = np.arange(0.01, 0.51, 0.01)
    best_alpha = 0.05
    best_rmse = np.inf

    for alpha in alphas:
        errors_sq = []
        for stock_id in target_vars.index.get_level_values("stock_id").unique():
            # subset for this stock, sorted by time_id
            idx = target_vars.index.get_level_values("stock_id") == stock_id
            stock_rv = rv_known[idx].sort_index(level="time_id")
            stock_target = target_vars[idx].sort_index(level="time_id")
            # Need at least 2 points to form one forecast
            if len(stock_rv) < 2:
                continue
            values = stock_rv.values
            # EWMA recursion
            s = values[0]
            # Forecast from second observation onward, error = (forecast - actual)^2
            for i in range(1, len(values)):
                forecast = s  # predict next variance as current smoothed value
                errors_sq.append((forecast - stock_target.values[i]) ** 2)
                s = alpha * values[i] + (1 - alpha) * s
        rmse = np.sqrt(np.mean(errors_sq)) if errors_sq else np.inf
        if rmse < best_rmse:
            best_rmse = rmse
            best_alpha = alpha
    return best_alpha

# ------------------------------------------------------------------
# 4. Predict for unseen (stock_id, time_id) using tuned EWMA
# ------------------------------------------------------------------
def predict_YOUR_MODEL_HERE(df: pd.DataFrame, prediction_index: pd.MultiIndex) -> pd.Series:
    """
    Predict variance for each (stock_id, time_id) in prediction_index.
    - Uses EWMA over known targets to forecast future variance.
    - Falls back to stock‑wise mean variance if no history exists.
    """
    global target_vars  # already loaded in the notebook

    # --- 4a. Compute realized variance for all known target pairs ---
    known_idx = target_vars.index
    rv_known = compute_rv(df, known_idx)  # same index as target_vars

    # --- 4b. Tune α (if not already done) ---
    # To avoid re‑running each call, you could cache the result outside.
    # For simplicity we tune every time – it's fast for typical data sizes.
    best_alpha = tune_alpha(target_vars, rv_known)

    # --- 4c. Build per‑stock EWMA state after each known time_id ---
    # stock_last_state: after processing all known time_ids, dict stock_id -> smoothed value
    # stock_state_dict: stock_id -> {time_id: smoothed value after including that time_id}
    stock_last_state = {}
    stock_state_dict = {}   # for interpolation
    for stock_id in rv_known.index.get_level_values("stock_id").unique():
        series = rv_known.loc[stock_id].sort_index()
        times = series.index.get_level_values("time_id")  # already sorted
        vals = series.values
        s = vals[0]
        state_after = {times[0]: s}
        for i in range(1, len(vals)):
            s = best_alpha * vals[i] + (1 - best_alpha) * s
            state_after[times[i]] = s
        stock_last_state[stock_id] = s
        stock_state_dict[stock_id] = state_after

    # Fallback: global mean of all known targets (or RV)
    global_mean = target_vars.mean() if not target_vars.empty else 0.0

    # --- 4d. Generate predictions ---
    preds = {}
    prediction_index = prediction_index.sort_values()  # ensure order
    for stock_id, time_id in prediction_index:
        if stock_id in stock_state_dict:
            known_times = stock_state_dict[stock_id]
            # Find the largest known time_id <= time_id
            earlier = [t for t in known_times.keys() if t < time_id]
            if earlier:
                last_known = max(earlier)
                pred = known_times[last_known]   # state after that time
            else:
                # No earlier known time – use the first known state
                pred = list(known_times.values())[0]
                
            preds[(stock_id, time_id)] = pred
        else:
            # Stock never seen in training
            # Use global mean of target_vars
            preds[(stock_id, time_id)] = global_mean

    return pd.Series(preds, name="predicted_variance")

def cost():
    predicted_vars = predict_YOUR_MODEL_HERE(df, known_prediction_index)
    return score(predicted_vars, target_vars, plot=True, log_axes=True)

cost()
````

````python
def predict_spread_weighted_var(df: pd.DataFrame, prediction_index: pd.MultiIndex) -> pd.Series:
    """
    Model 2: Liquidity-Weighted Variance.
    Weights price jumps heavier if they occur during times of low liquidity (wide spreads).
    """
    def _spread_var(group: pd.DataFrame) -> float:
        if group.empty:
            return 0.0
        
        log_returns = group["log_return"].dropna()
        if log_returns.empty:
            return 0.0
        
        # Calculate the Bid-Ask Spread for each row
        # Formula: (Ask - Bid) / Weighted Average Price
        spread = (group["ask_price"] - group["bid_price"]) / group["weighted_average_price"]
        
        # Match the spread data to the rows where we have log_returns
        aligned_spread = spread.loc[log_returns.index]
        
        # Weight the squared returns by the spread size
        weighted_sq_returns = (log_returns ** 2) * aligned_spread
        
        # Normalize the result so the scale matches normal variance
        mean_spread = aligned_spread.mean()
        if mean_spread > 0:
            return float(weighted_sq_returns.sum() / mean_spread)
        else:
            # Fallback if spread is somehow exactly 0
            return float((log_returns ** 2).sum())

    print("Grouping and calculating spread-weighted variance...")
    gdf = df.groupby(["stock_id", "time_id"], sort=True)
    all_vars = gdf.apply(lambda grp: _spread_var(grp), include_groups=False)
    
    return all_vars.reindex(prediction_index).fillna(0.0)

def cost():
    predicted_vars = predict_spread_weighted_var(df, known_prediction_index)
    return score(predicted_vars, target_vars, plot=True, log_axes=True)

cost()
````

### Explanatory Notes

Below are some explanatory notes about the stock data used in this workshop. Can you use this information to make a better variance prediction?

#### The Order Book

Traders in a stock market place _orders_, each of which is a request to buy or sell a certain number of shares in a stock at a certain price. Taken together, the current orders can be viewed as an _Order Book_ which shows the total number of shares (a.k.a. volume) that traders are willing to buy or sell at each price level. For example:

| Bid Volume | Price | Sell Volume |
| ---------: | :---: | :---------- |
|            |  151  |  196        |
|            |  150  |  189        |
|            |  149  |  148        |
|            |  148  |  221        |
|       251  |  147  |             |
|       321  |  146  |             |
|       300  |  145  |             |
|        20  |  144  |             |

In this example, there are traders willing to sell at a price of 148 and the total number of shares they are willing to sell at that price is 221. Similarly there are traders willing to buy up to 321 shares at a price of 146.

We number the buy (bid) prices from highest to lowest and the sell (ask) prices from lowest to highest. In our data, `bid_price` is the highest buy price in the order book and `ask_price` is the lowest sell price.


#### Weighted Average Price

We can use the order book to get a sense of the _current_ price by taking an average of the prices at which traders are willing to trade, weighted by the volume at each price level.

A simple weighted average would, however, be misleading. To understand why, suppose there were many buyers and few sellers. In that case the buy (lower) prices would have greater weight than the sell (higher) prices, pulling the weighted average down. But if there are more buyers than sellers, we would expect the price to go up, not down.

Therefore, the formula we use for the average price is:

$$
\frac{\mathit{BidPrice}\times\mathit{AskSize} + \mathit{AskPrice}\times\mathit{BidSize}}{\mathit{BidSize} + \mathit{AskSize}}
$$


#### Returns

Traders naturally want to know whether they are making a profit or a loss on their investments. The _return_ over a period of time is simply the price at the end of that period divided by the price at the beginning.

Returns are widely used in finance, however _log returns_ are preferred whenever some mathematical modelling is required. If $t_1$ and $t_2$ are two points in time, then the _log return_ from $t_1$ to $t_2$ is:

$$
r_{t_1, t_2} = \log \left(\frac{\mathit{WeightedAveragePrice}_{t_2}}{\mathit{WeightedAveragePrice}_{t_1}} \right)
$$

Log returns have several advantages including, for example:
- they are additive across time $r_{t_1, t_2} + r_{t_2, t_3} = r_{t_1, t_3}$
- regular returns cannot go below -100%, while log returns are not bounded.

Note: in most of the notebook we write $r_{t_1, t_2}$ as $r_{t}$ for readability
