```python
from pathlib import Path
import pandas as pd

REPO = None
for _p in [Path.cwd(), *Path.cwd().parents]:
    if (_p / "data" / "plotting.py").is_file():
        REPO = _p
        break
if REPO is None:
    raise RuntimeError("Could not find repo root (data/plotting.py). Run from opti-ff or a subfolder.")

DATA_DIR = REPO / "data" / "csv"
TAIL = 10_000

paths = sorted(
    p for p in DATA_DIR.glob("*_strategy_market_data.csv")
    if p.is_file() and ".ipynb_checkpoints" not in p.parts
)

if not paths:
    raise FileNotFoundError(f"No *_strategy_market_data.csv in {DATA_DIR}")

dfs = []
for path in paths:
    inst = path.name.replace("_strategy_market_data.csv", "")
    chunk = pd.read_csv(path, usecols=["timestamp", "mid"], encoding="utf-8-sig")
    chunk["mid"] = pd.to_numeric(chunk["mid"], errors="coerce")
    chunk["timestamp"] = pd.to_numeric(chunk["timestamp"], errors="coerce")
    chunk = chunk.dropna(subset=["timestamp", "mid"]).tail(TAIL)
    chunk = chunk.rename(columns={"mid": inst})
    dfs.append(chunk[["timestamp", inst]])

prices_wide = dfs[0]
for d in dfs[1:]:
    prices_wide = prices_wide.merge(d, on="timestamp", how="inner")

prices_wide = prices_wide.sort_values("timestamp").reset_index(drop=True)
instrument_cols = [c for c in prices_wide.columns if c != "timestamp"]
prices_wide = prices_wide.dropna(subset=instrument_cols, how="any")

print(f"Instruments: {len(instrument_cols)}")
print(f"Rows after inner-join + dropna: {len(prices_wide)}")
prices_wide.tail()
```

```python
corr_matrix = prices_wide[instrument_cols].corr()
corr_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_matrix,
    cmap="coolwarm",
    center=0.0,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.75},
)
plt.title("Correlation matrix (mid prices, aligned timestamps)")
plt.tight_layout()
plt.show()
```

```python
pairs = []
cols = list(corr_matrix.columns)
for i, a in enumerate(cols):
    for b in cols[i + 1 :]:
        pairs.append((a, b, float(corr_matrix.loc[a, b])))

top_corr = (
    pd.DataFrame(pairs, columns=["instrument_a", "instrument_b", "correlation"])
    .assign(abs_corr=lambda d: d["correlation"].abs())
    .sort_values("abs_corr", ascending=False)
    .reset_index(drop=True)
)

# Top 30 strongest linear relationships (by absolute correlation)
top_corr.head(10)
```

```python
# NVDA / NVDA_DUAL spread (expects ``prices_wide`` from earlier cells; alias as ``data_wide``).
data_wide = prices_wide

nvda = data_wide["NVDA"].copy()[-600:]
nvda_dual = data_wide["NVDA_DUAL"].copy()[-600:]
spread_nvda = nvda - nvda_dual
spread_nvda.name = "NVDA_minus_NVDA_DUAL"

import matplotlib.pyplot as plt

x_ev = range(len(spread_nvda))
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x_ev, spread_nvda.to_numpy(), label="NVDA - NVDA_DUAL", linewidth=1.0)
ax.set_xlabel("Event number")
ax.set_ylabel("Mid spread")
ax.set_title("NVDA vs NVDA_DUAL: price spread")
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import statsmodels.tsa.stattools as ts
```

```python
result = ts.coint(nvda, nvda_dual)

cointegration_t_statistic = result[0]
p_val = result[1]
critical_values_test_statistic_at_1_5_10 = result[2]

print('We want the P val < 0.05 (meaning that cointegration exists)')
print('P value for the augmented Engle-Granger two-step cointegration test is', p_val)
```

```python
from statsmodels.tsa.stattools import adfuller

# With time series, stationary inputs are easier to model; ADF tests unit root (null: non-stationary).
NVDA_ADF = adfuller(nvda)
print('P value for the Augmented Dickey-Fuller Test is', NVDA_ADF[1])
NVDA_DUAL_ADF = adfuller(nvda_dual)
print('P value for the Augmented Dickey-Fuller Test is', NVDA_DUAL_ADF[1])
Spread_ADF = adfuller(spread_nvda)
print('P value for the Augmented Dickey-Fuller Test is', Spread_ADF[1])
ratio_nvda = nvda / nvda_dual
Ratio_ADF = adfuller(ratio_nvda.dropna())
print('P value for the Augmented Dickey-Fuller Test is', Ratio_ADF[1])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6), dpi=200)
ratio = nvda / nvda_dual
x_ev = range(len(ratio))
plt.plot(x_ev, ratio.to_numpy(), label="Price Ratio (NVDA / NVDA_DUAL)")
plt.axhline(ratio.mean(), color="red")
plt.xlabel("Event number")
plt.legend()
plt.title("Price Ratio between NVDA and NVDA_DUAL")
plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=200)
ratio = nvda / nvda_dual
df_zscore = (ratio - ratio.mean()) / ratio.std()
x_ev = range(len(df_zscore))
plt.plot(x_ev, df_zscore.to_numpy(), label="Z Scores")
plt.xlabel("Event number")
plt.axhline(df_zscore.mean(), color="black")
plt.axhline(1.0, color="red")
plt.axhline(1.25, color="red")
plt.axhline(-1.0, color="green")
plt.axhline(-1.25, color="green")
plt.legend(loc="best")
plt.title("Z score of Ratio of NVDA to NVDA_DUAL")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# --- scikit-learn linear regression on event index (80/20 time-ordered split) ---
# ``ratio`` = NVDA / NVDA_DUAL. Model: ratio ≈ β0 + β1 · (event number); intercept is learned.
# Buy / sell from residual cups vs peaks vs the learned trend; train σ sets thresholds.

ratio = (nvda / nvda_dual).copy().astype(float)
y = ratio.to_numpy(dtype=float)
x_ev = np.arange(len(y), dtype=float)
n = len(y)
if n < 10:
    raise ValueError("Need at least 10 rows for an 80/20 split and regression.")

# One feature: event number (sklearn adds intercept by default).
X = x_ev.reshape(-1, 1)
n_train = max(2, int(0.8 * n))
X_train, y_train = X[:n_train], y[:n_train]

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X)

residual = y - y_pred
std_train = float(np.std(residual[:n_train], ddof=1))
if std_train < 1e-12:
    std_train = 1e-12

k = 1.0
buy_mask = residual < -k * std_train
sell_mask = residual > k * std_train

zscore_20_5 = pd.Series(residual / std_train, index=ratio.index)

plt.figure(figsize=(10, 5), dpi=200)
plt.plot(x_ev, y, label="Ratio (NVDA / NVDA_DUAL)", alpha=0.85)
plt.plot(x_ev, y_pred, label="sklearn LinearRegression (fit on first 80%)", linewidth=2)
plt.axvline(n_train - 0.5, color="gray", linestyle=":", label="80 / 20 split")
plt.xlabel("Event number")
plt.ylabel("Ratio")
plt.title("NVDA / NVDA_DUAL ratio vs sklearn linear regression on event number")
plt.legend(loc="best")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5), dpi=200)
plt.plot(x_ev, residual / std_train, label="Residual / train σ")
plt.axhline(0.0, color="black", linewidth=1)
plt.axhline(k, color="red", linestyle="--", label=f"sell if residual > +{k}σ")
plt.axhline(-k, color="green", linestyle="--", label=f"buy if residual < −{k}σ")
plt.axvline(n_train - 0.5, color="gray", linestyle=":")
plt.xlabel("Event number")
plt.ylabel("Standardized residual")
plt.title("Cups (negative) vs peaks (positive) relative to the linear model")
plt.legend(loc="best")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5), dpi=200)
plt.plot(x_ev, y, label="Ratio", color="tab:blue", alpha=0.8)
plt.plot(x_ev, y_pred, label="Model (sklearn)", color="tab:orange", linewidth=2)
plt.scatter(
    x_ev[buy_mask],
    y[buy_mask],
    color="green",
    marker="^",
    s=50,
    zorder=5,
    label="Buy (cup: below model)",
)
plt.scatter(
    x_ev[sell_mask],
    y[sell_mask],
    color="red",
    marker="v",
    s=50,
    zorder=5,
    label="Sell (peak: above model)",
)
plt.axvline(n_train - 0.5, color="gray", linestyle=":", label="80 / 20 split")
plt.xlabel("Event number")
plt.ylabel("Ratio")
plt.title("sklearn LinearRegression buy/sell from residual vs train σ")
plt.legend(loc="best")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Requires ``ratio`` and ``zscore_20_5`` from the prior cell (rolling MA / z-score).
plt.figure(figsize=(8, 6), dpi=200)
ratio.plot()
buy = ratio.copy()
sell = ratio.copy()
buy[zscore_20_5 > -1] = 0
sell[zscore_20_5 < 1] = 0
buy.plot(color="g", linestyle="None", marker="^")
sell.plot(color="r", linestyle="None", marker="^")
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, ratio.min(), ratio.max()))
plt.legend(["Ratio", "Buy Signal", "Sell Signal"])
plt.xlabel("Event number")
plt.title("NVDA / NVDA_DUAL ratio with z-score buy/sell markers")
plt.show()
```
